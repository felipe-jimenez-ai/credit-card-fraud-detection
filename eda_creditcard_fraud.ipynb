```python
# Detecting outliers with boxplots and interquartile range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
```

## ⚖️ Balance Classes

99% legitimate and 1% fraud. This creates a class imbalance.

This imbalance is a problem because:

A model that always predicts "no fraud" will have high accuracy, but will be useless in detecting fraud.
Algorithms learn best when there are balanced examples of each class.

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

In Python, the library offers easy methods for balancing classes, such as:imbalanced-learn

RandomOverSampler: Duplicates examples of the minority class.
SMOTE: Creates synthetic examples based on close neighbors.

The most common techniques for balancing classes are:

* **Oversampling:** Create or copy more examples of the minority class (fraud).
* **Undersampling:** Reduce the number of examples of the majority class (legitimate).
* **SMOTE:** A method that creates new synthetic examples for the minority class using similar neighbors

```python
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
```

## Download Cleaned Creditcard Dataset

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler

# Separar variables y etiquetas
X = df.drop('Class', axis=1)
y = df['Class']

# Aplicar balanceo
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)

# Guardar el dataset limpio
df_cleaned = pd.concat([X_res, y_res], axis=1)
df_cleaned.to_csv("creditcard_cleaned.csv", index=False)
```
